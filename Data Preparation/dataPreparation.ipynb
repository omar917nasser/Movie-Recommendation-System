```python
import pandas as pd
```

```python
# Sample datasets (already cleaned)
actor_df = pd.read_csv(r"D:\Coding\Code\Python\ML\Data\IMDB\Cleaned Data\FinalActorBasedRecomendation.csv")  # tconst, nconst, primaryName, etc.
actor_df
```

```python
director_df = pd.read_csv(r"D:\Coding\Code\Python\ML\Data\IMDB\Cleaned Data\FinalDirectorBasedRecomendation.csv")  # tconst, directors, primaryName, etc.
director_df
```

```python
genre_df = pd.read_csv(r"D:\Coding\Code\Python\ML\Data\IMDB\Cleaned Data\FinalGenreBasedRecomendation.csv")  # tconst, genres
genre_df
```

```python
rating_df = pd.read_csv(r"D:\Coding\Code\Python\ML\Data\IMDB\Cleaned Data\FinalPopularityBasedRecomendation.csv")  # tconst, averageRating, numVotes
rating_df
```

```python
combined_df = pd.merge(actor_df[['tconst', 'nconst']], director_df[['tconst', 'directors']], on='tconst', how='inner')
combined_df
```

```python
combined_df = pd.merge(combined_df, genre_df[['tconst', 'genres']], on='tconst', how='inner')
combined_df
```

```python
combined_df = pd.merge(combined_df, rating_df[['tconst', 'popularity_score']], on='tconst', how='inner')
combined_df
```

```python
combined_df['genres'] = combined_df['genres'].apply(lambda x: x.strip("[]").replace("'", "").split(",") if isinstance(x, str) else [])
combined_df
```

```python
combined_df.to_csv(r"D:\Coding\Code\Python\ML\Data\IMDB\Cleaned Data\FinalCombined.csv")
```
