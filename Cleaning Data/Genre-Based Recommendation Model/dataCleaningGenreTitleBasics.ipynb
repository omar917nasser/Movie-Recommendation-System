```python
import pandas as pd
from sklearn.preprocessing import MultiLabelBinarizer
```

```python
title_basics = pd.read_csv(r"D:\Coding\Code\Python\ML\Data\IMDB\title.basics.tsv", sep="\t")
print("\nTitle.Akas")
title_basics
```

```python
title_basics.replace("\\N", pd.NA , inplace=True)
title_basics.isna().sum()
```

```python
title_basics_cleaned = title_basics.drop(columns=['isAdult', 'originalTitle', 'endYear','startYear','runtimeMinutes'])
title_basics_cleaned
```

```python
title_basics_cleaned = title_basics_cleaned.dropna(subset=['genres','primaryTitle'])
title_basics_cleaned.isna().sum()
```

```python
title_basics_cleaned['genres'] = title_basics_cleaned['genres'].apply(lambda x: x.split(',') if isinstance(x, str) else [])
title_basics_cleaned
```

```python
title_basics_cleaned['genres'] = title_basics_cleaned['genres'].apply(lambda x: [genre.lower() for genre in x])
title_basics_cleaned
```

```python
title_basics_cleaned.to_csv(r"D:\Coding\Code\Python\ML\Data\IMDB\Cleaned Data\FinalGenreBasedRecomendation.csv")
```
