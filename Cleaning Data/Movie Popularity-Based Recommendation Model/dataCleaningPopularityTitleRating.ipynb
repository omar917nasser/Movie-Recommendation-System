```python
import pandas as pd
```

```python
title_ratings = pd.read_csv(r"D:\Coding\Code\Python\ML\Data\IMDB\title.ratings.tsv", sep="\t")
print("\nTitle.Ratings")
title_ratings
```

```python
title_ratings.replace("\\N", pd.NA , inplace=True)
title_ratings.isna().sum()
```

```python
title_rating_cleaned = title_ratings[title_ratings['numVotes'] >= 10]
title_rating_cleaned
```

```python
C = title_rating_cleaned['averageRating'].mean()
m = 10  # Minimum number of votes required

title_rating_cleaned['popularity_score'] = (
    (title_rating_cleaned['numVotes'] * title_rating_cleaned['averageRating']) +
    (m * C)
) / (title_rating_cleaned['numVotes'] + m)
title_rating_cleaned
```

```python
title_rating_cleaned = title_rating_cleaned.sort_values(by='popularity_score', ascending=False)
title_rating_cleaned
```

```python
title_rating_cleaned.to_csv(r"D:\Coding\Code\Python\ML\Data\IMDB\Cleaned Data\FinalPopularityBasedRecomendation.csv")
```
