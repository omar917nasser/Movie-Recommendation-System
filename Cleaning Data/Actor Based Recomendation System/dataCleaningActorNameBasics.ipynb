```python
import pandas as pd
```

```python
name_basics = pd.read_csv(r"D:\Coding\Code\Python\ML\Data\IMDB\name.basics.tsv", sep="\t")
print("Name.Basics")
name_basics
```

```python
name_basics.replace("\\N", pd.NA , inplace=True)
name_basics.isna().sum()
```

```python
name_basics_cleaned = name_basics.dropna(subset=['primaryName', 'primaryProfession'])
name_basics_cleaned.isna().sum()
```

```python
name_basics_cleaned.drop(columns=['birthYear','deathYear','knownForTitles'], inplace=True)
name_basics_cleaned
```

```python
name_basics_cleaned.isna().sum()
```

```python
name_basics_cleaned_actor = name_basics_cleaned[name_basics_cleaned['primaryProfession'].str.contains('actor|actress', na=False)]
name_basics_cleaned_actor
```

```python
name_basics_cleaned_actor.to_csv(r"D:\Coding\Code\Python\ML\Data\IMDB\Cleaned Data\name_basics_cleaned_actor.csv")
```
