```python
import pandas as pd
```

```python
title_crew = pd.read_csv(r"D:\Coding\Code\Python\ML\Data\IMDB\title.crew.tsv", sep="\t")
print("\nTitle.Crew")
title_crew
```

```python
title_crew.replace("\\N", pd.NA , inplace=True)
title_crew.isna().sum()
```

```python
title_crew_cleaned = title_crew.drop(columns=['writers'])
title_crew_cleaned
```

```python
title_crew_cleaned.dropna(subset=['directors'], inplace=True)
title_crew_cleaned.isna().sum()
```

```python
title_crew_cleaned
```

```python
rows_with_commas = title_crew_cleaned[title_crew['directors'].str.contains(',', na=False)]
rows_with_commas
```

```python
title_crew_cleaned = title_crew_cleaned.assign(directors=title_crew_cleaned['directors'].str.split(',')).explode('directors')
```

```python
rows_with_commas = title_crew_cleaned[title_crew['directors'].str.contains(',', na=False)]
rows_with_commas
```
